# 2. Concat

#### ID 221부터 230의 학생 정보 

```python
SI_v2 = {'ID' : [i for i in range(221, 231)],
         'Gender' : ['M', 'F', 'M', 'M', 'M', 'M', 'F', 'F', 'F', 'F'],
         'Class' : ['C', 'D', 'C', 'B', 'B', 'D', 'A', 'E', 'A', 'E'],     
         'Score' : [84, 92, 86, 77, 83, 75, 91, 94, 85, 98],
         'Country' : ['China', 'USA', 'Korea', 'China', 'English','USA', 'Japan', 'Korea', 'Japan', 'Korea'],
         'City_Id' : [8, 12, 13, 17, 22, 20, 23, 10, 23, 14]
         }

SI_v2 = DataFrame(SI_v2)
SI_v2
```
