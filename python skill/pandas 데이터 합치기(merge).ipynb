```python
from pandas import DataFrame
import pandas as pd
```

```python
df1 = {'id' : [211, 214, 220, 253],
       'class' : ['A', 'B', 'C', 'A'],
       'score' : [70, 80, 90, 84]}
df1 = DataFrame(df1)
df1
```

```python
df2 = {'city' : ['Seoul', 'Tokyo', 'Beijing', 'Busan', 'Seoul', 'NewYork', 'Paris', 'Busan', 'Paris', 'Seoul'],
       'id' : [210, 211, 212, 213, 214, 215, 216, 220, 223, 253],
       'Y/N' : ['Y', 'N', 'N','N', 'Y', 'N', 'Y', 'Y', 'N', 'Y']}

df2 = DataFrame(df2)
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on = 'id')
```
