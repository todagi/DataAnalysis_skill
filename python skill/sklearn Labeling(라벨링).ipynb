```python
from pandas import DataFrame
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
# si = student infromation

si = {'ID' : [i for i in range(201, 221)],
      'Gender' : ['M', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'M', 'M',
                   'F', 'F', 'F', 'F', 'F', 'M', 'F', 'F', 'M', 'M'],
      'Class' : ['D', 'E', 'A', 'B', 'B', 'C', 'E', 'C', 'D', 'E',
                 'E', 'D', 'C', 'B', 'A', 'A', 'C', 'D', 'B', 'A'],     
      'Score' : [85, 78, 84, 79, 80, 99, 86, 91, 83, 87,
                 93, 75, 77, 94, 72, 74, 65, 95, 100, 76],
      'Country' : ['Japan', 'USA', 'China', 'China', 'USA', 'Japan', 'USA', 'USA', 'Korea', 'Korea',
                   'Korea', 'China', 'English', 'Korea', 'Korea', 'Korea', 'Japan', 'USA', 'Korea', 'Japan'],
      'City_Id' : [15, 12, 8, 17, 19, 16, 20, 20, 13, 14, 10, 8, 9, 14, 11, 12, 10, 11, 10, 15]
      }

si = DataFrame(si)
si
```

#### 성별 'M'과 'F' 문자열 데이터를 숫자(카테고리까지)로 바꾸기

```python
si.Gender.value_counts()
```

```python
label = LabelEncoder()
```

```python
label.fit(si['Gender'])
```

```python
label.classes_
```

```python
label.transform(si['Gender'])
```

```python
si['Gender'] = label.transform(si['Gender'])
```

```python
si
```

```python
si['Gender'].value_counts()
```

#### Labeling 결과

```python
성별(Gender)     라벨링    명수           

F(Female)          (1)      10      
M(Male)            (0)      10 
```
